```python
import pandas as pd
train = pd.read_csv("../../data/kaggle/house_prices/train.csv")
test = pd.read_csv("../../data/kaggle/house_prices/test.csv")
sample =  pd.read_csv("../../data/kaggle/house_prices/sample_submission.csv")
```

```python
# 現在の方針
# 列が非常に多く、モデル構築の為に欠損を補おうにもデータの全貌が見えない状態
# → 扱いやすい数値列に列を絞る

numeric_train = train.select_dtypes(include=["int64", "float64"])
numeric_train.head()
```

```python
print(len(numeric_train))
numeric_train.isnull().sum().sort_values(ascending=False)
# sort_valuesを使用し、データを並べ替え
# ascending=は昇順か降順か　降順にすることで欠損値を網羅
```

```python
LotFrontage_median = numeric_train["LotFrontage"].median()
numeric_train["LotFrontage"] = numeric_train["LotFrontage"].fillna(LotFrontage_median)

GarageYrBlt_median = numeric_train["GarageYrBlt"].median()
numeric_train["GarageYrBlt"] = numeric_train["GarageYrBlt"].fillna(GarageYrBlt_median)

MasVnrArea_median = numeric_train["MasVnrArea"].median()
numeric_train["MasVnrArea"] = numeric_train["MasVnrArea"].fillna(MasVnrArea_median)
# 欠損値を中央値で補完

numeric_train.isnull().sum().sort_values(ascending=False)
```

```python
import numpy as np
from sklearn import linear_model
import sklearn.model_selection
# ライブラリのインポート

X = numeric_train
y = np.log1p(numeric_train["SalePrice"])
# SalePriceをlog変換

# logの概念
# log₄64 = 3 ⇔ 4³ = 64

# np.log(Ｘ)におけるlog
# logₑ(Ｘ)=Ｙ ⇔ eのＹ乗= Ｘ

# eは固定の数(2.718281828…)
# 機械学習、微分、確率で都合が良い数字になっている

# np.log()をするとどうなるか
# → 倍率が同じなら、log の差は一定になる

# 例)
# Ｘ	　log(Ｘ)
# 10	　約2.30
# 100	　約4.61
# 1000	　約6.91
```

```python
del X["SalePrice"]

X_train,X_test,y_train,y_test = sklearn.model_selection.train_test_split(X,y,random_state=0)
# Xが教材　ｙが答え　trainが学習用　testが評価用
model = linear_model.LinearRegression()
model.fit(X_train,y_train)
# x₁ x₂	y
# 1  1  3
# 2	 1	5
# 3	 2	8

# ｙ=w1x1+w2x2+b
# で、最適なw1,w2、を学習している
# 具体的な計算は(XtX)−1Xty ← 長期的に学習

# 各特徴量が価格にどれくらい寄与するかを係数として学習し、
# それらを足し合わせて価格を推定している
# 価格は分布が歪むため log 変換し、誤差が相対的に均等になるようにしている

print(model.score(X_test,y_test))
print(model.score(X_train,y_train))
# 評価用での精度が約75%,学習用での精度が約89% → 明らかに過学習/チューニングが必要
```

```python
# Ridgeを使用し、係数が大きくなりすぎたら、ペナルティを与え過学習を抑制することで、
# より洗練されたモデルを構築することを目指したが、元の数値が最も高かった
```

```python
# 現在の方針　
# モデルの作成完了/test(実際の予測に使うデータ)が加工前なので、加工
numeric_test = test.select_dtypes(include=["int64", "float64"])
numeric_test.isnull().sum().sort_values(ascending=False)
```

```python
LotFrontage_median = numeric_test["LotFrontage"].median()
numeric_test["LotFrontage"] = numeric_test["LotFrontage"].fillna(LotFrontage_median)

GarageYrBlt_median = numeric_test["GarageYrBlt"].median()
numeric_test["GarageYrBlt"] = numeric_test["GarageYrBlt"].fillna(GarageYrBlt_median)

MasVnrArea_median = numeric_test["MasVnrArea"].median()
numeric_test["MasVnrArea"] = numeric_test["MasVnrArea"].fillna(MasVnrArea_median)

BsmtHalfBath_median = numeric_test["BsmtHalfBath"].median()
numeric_test["BsmtHalfBath"] = numeric_test["BsmtHalfBath"].fillna(BsmtHalfBath_median)

BsmtFullBath_median = numeric_test["BsmtFullBath"].median()
numeric_test["BsmtFullBath"] = numeric_test["BsmtFullBath"].fillna(BsmtFullBath_median)

GarageCars_median = numeric_test["GarageCars"].median()
numeric_test["GarageCars"] = numeric_test["GarageCars"].fillna(GarageCars_median) 

GarageArea_median = numeric_test["GarageArea"].median()
numeric_test["GarageArea"] = numeric_test["GarageArea"].fillna(GarageArea_median) 

BsmtFinSF2_median = numeric_test["BsmtFinSF2"].median()
numeric_test["BsmtFinSF2"] = numeric_test["BsmtFinSF2"].fillna(BsmtFinSF2_median) 

BsmtUnfSF_median = numeric_test["BsmtUnfSF"].median()
numeric_test["BsmtUnfSF"] = numeric_test["BsmtUnfSF"].fillna(BsmtUnfSF_median) 

TotalBsmtSF_median = numeric_test["TotalBsmtSF"].median()
numeric_test["TotalBsmtSF"] = numeric_test["TotalBsmtSF"].fillna(TotalBsmtSF_median)  

BsmtFinSF1_median = numeric_test["BsmtFinSF1"].median()
numeric_test["BsmtFinSF1"] = numeric_test["BsmtFinSF1"].fillna(BsmtFinSF1_median) 
# 中央値で補完
# おそらく、もっとスマートな方法があるので、improvedで要改善


numeric_test.isnull().sum().sort_values(ascending=False)
```

```python
pred_log = model.predict(numeric_test)
pred = np.expm1(pred_log)
# logを元に戻す
```

```python
submission = pd.DataFrame({"Id": test["Id"],"SalePrice": pred})
submission.to_csv("../../outputs/kaggle_house_price/submission.csv", index=False)
```
